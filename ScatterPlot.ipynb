```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)  # Random x values
y = np.random.rand(50)  # Random y values

plt.scatter(x, y)
plt.title('Basic Scatter Plot')
plt.xlabel('X-axis',fontsize=15)
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  # Random colors
sizes = 1000 * np.random.rand(50)  # Random sizes

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title('Scatter Plot with Custom Colors and Sizes')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)  # Variable for color map

plt.scatter(x, y, c=z, cmap='viridis', alpha=0.7)
plt.colorbar(label='Color Scale')
plt.title('Scatter Plot with Color Map')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
x = np.random.rand(50)
y = 2 * x + np.random.rand(50)  # Linear relationship with some noise

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
plt.scatter(x, y, label='Data points',marker='*',edgecolor='green',linewidth=4)
plt.plot(x, slope * x + intercept, color='red', label='Regression Line')
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.rand(50)

plt.scatter(x1, y1, c='blue', label='Group 1')
plt.scatter(x2, y2, c='red', label='Group 2')
plt.title('Scatter Plot with Multiple Groups')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
x = np.random.rand(10)
y = np.random.rand(10)

plt.scatter(x, y)
for i in range(len(x)):
    plt.annotate(f'Point {i+1}', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.title('Scatter Plot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
ax.set_title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, alpha=0.4)  # Set alpha for transparency
plt.title('Scatter Plot with Transparency')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, marker='s', color='green')  # Square marker
plt.title('Scatter Plot with Custom Marker Style')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
