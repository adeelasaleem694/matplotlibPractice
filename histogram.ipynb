```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.randn(100)

# Plot histogram
plt.hist(data, bins=10, color='blue', edgecolor='black')
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

# Plot histogram with density
plt.hist(data, bins=20, color='purple', edgecolor='black',density=True)
plt.title('Density Histogram')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  # Shifted dataset

# Plot histograms
plt.hist(data1, bins=20, alpha=0.7, label='Dataset 1', edgecolor='black', density=True)
plt.hist(data2, bins=20, alpha=0.7, label='Dataset 2', edgecolor='black', density=True)
plt.title('Comparing Two Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
data = np.random.randint(1, 100, 50)

# Plot horizontal histogram
plt.hist(data, bins=10, color='orange', edgecolor='black', orientation='horizontal')
plt.title('Horizontal Histogram')
plt.xlabel('Frequency')
plt.ylabel('Value')
plt.show()
```

```python
data = np.random.randn(1000)

# Plot cumulative histogram
plt.hist(data, bins=20, color='magenta', edgecolor='black', cumulative=True)
plt.title('Cumulative Histogram')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 3

# Subplots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(data1, bins=20, color='blue', edgecolor='black')
plt.title('Dataset 1')

plt.subplot(1, 2, 2)
plt.hist(data2, bins=20, color='green', edgecolor='black')
plt.title('Dataset 2')

plt.tight_layout()
plt.show()
```

```python
data = np.random.exponential(scale=2, size=1000)

# Plot histogram with log scale
plt.hist(data, bins=20, color='brown', edgecolor='black')
plt.yscale('log')
plt.title('Histogram with Log Scale')
plt.xlabel('Value')
plt.ylabel('Frequency (log scale)')
plt.show()
```
