```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = np.sin(x)

# Create the plot
plt.plot(x, y, label='Sine Wave')  # Plot x vs. y
plt.title('Basic Line Plot')  # Add a title
plt.xlabel('X-axis')          # Label X-axis
plt.ylabel('Y-axis')          # Label Y-axis
plt.legend()                  # Show legend
plt.grid(True)                # Add a grid
plt.show()                    # Display the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 34]

# Customize figure size
plt.figure(figsize=(10, 5))  # Width: 10, Height: 6 (in inches)

# Create bar plot
plt.bar(categories, values, color='skyblue', width=0.6, edgecolor='black',linewidth=5,align='edge') #linestyle=':',alpha=0.5

# Add titles and labels with custom font sizes
plt.title('Bar Plot Example', fontsize=16, fontweight='bold')  # Title with font size
plt.xlabel('Categories', fontsize=14)  # X-axis label with font size
plt.ylabel('Values', fontsize=14)  # Y-axis label with font size

# Add grid
plt.grid(axis='x', linestyle='--', alpha=0.8)  # Grid only on y-axis
plt.grid(axis='y', linestyle='--', alpha=0.8) 
# Customize tick parameters
plt.xticks(fontsize=10)  # X-axis ticks font size
plt.yticks(fontsize=12)  # Y-axis ticks font size

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [23, 45, 56, 78, 34]  # Data for the first bar set
values2 = [35, 40, 60, 70, 50]  # Data for the second bar set

# Number of categories
x = np.arange(len(categories))  # Positions for the categories

# Bar width
bar_width = 0.4

# Create the plot
plt.figure(figsize=(10, 6))  # Customize figure size

# Plot the first set of bars
plt.bar(x - bar_width/2, values1, width=bar_width, label='Set 1', color='skyblue', edgecolor='black')

# Plot the second set of bars
plt.bar(x + bar_width/2, values2, width=bar_width, label='Set 2', color='orange', edgecolor='black')

# Add labels and title
plt.title('Combined Bar Graph ', fontsize=16, fontweight='bold')
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Add custom x-axis labels
plt.xticks(x, categories, fontsize=12)

# Add a legend
plt.legend(fontsize=12)

# Add a grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()

```

```python

# Data for the bar plot
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [23, 45, 56, 78, 34]  # Data for the bars

# Create the plot
plt.figure(figsize=(8, 6)) 

# Horizontal bar plot
plt.barh(categories, values, color='skyblue', edgecolor='black', height=0.5)

# Add labels and title
plt.title('Horizontal Bar Graph', fontsize=16, fontweight='bold')
plt.xlabel('Values', fontsize=14)
plt.ylabel('Categories', fontsize=14)

# Add a grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Data for two sets
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values1 = [23, 45, 56, 78, 34]
values2 = [35, 40, 60, 70, 50]

# Bar height
bar_height = 0.4

# Create positions for the bars
y = np.arange(len(categories))

plt.figure(figsize=(10, 6))  
# Plot first set of horizontal bars
plt.barh(y - bar_height/2, values1, height=bar_height, label='Set 1', color='skyblue', edgecolor='black')

# Plot second set of horizontal bars
plt.barh(y + bar_height/2, values2, height=bar_height, label='Set 2', color='orange', edgecolor='black')

# Add custom y-axis labels
plt.yticks(y, categories, fontsize=12)

# Add labels and title
plt.title('Combined Horizontal Bar Graph', fontsize=16, fontweight='bold')
plt.xlabel('Values', fontsize=14)
plt.ylabel('Categories', fontsize=14)

# Add legend
plt.legend(fontsize=12)

# Add a grid
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
